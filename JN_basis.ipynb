#### ТИПОВА СТРУКТУРА JN ДЛЯ МН

1. __УВЕДЕННЯ__ - повинно містити опис проблеми (контекст), цілі аналізу та питання на які треба знайти відповідь
2. __ПІДГОТОВКА ДАНИХ__ - включає контроль, очистку та збагачення (__ETL__)
3. __ПОПЕРЕДНІЙ АНАЛІЗ ДАНИХ (EDA)__ - виявленя різноманітних властивостей набора та встановлення базових показчиків (_base line_)
4. __МОДЕЛЮВАННЯ__ - вибір, налагодження та оптимізація моделей
5. __ВИСНОВКИ__ - ілюстрація досягнутих результатів та відповіді на поставлені в п.1 питання
